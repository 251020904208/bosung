```
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))
c = int(input("Nhập số thứ ba: "))
max_num = max(a, b, c)
print("Số lớn nhất là:", max_num)
```

```
year = int(input("Nhập vào một năm: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "là năm nhuận")
else:
    print(year, "không phải là năm nhuận")
```

```
n = int(input("Nhập số nguyên dương n: "))
tong = 0
for i in range(1, n + 1):
    tong += i
print("Tổng từ 1 đến", n, "là:", tong)
```

```
am = duong = bang0 = 0

print("Nhập 10 số nguyên:")
for i in range(10):
    x = int(input(f"Số thứ {i+1}: "))
    if x > 0:
        duong += 1
    elif x < 0:
        am += 1
    else:
        bang0 += 1
print("Số dương:", duong)
print("Số âm:", am)
print("Số bằng 0:", bang0)
```

```
n = int(input("Nhập một số từ 1–9: "))
if 1 <= n <= 9:
    print(f"Bảng cửu chương {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")
else:
    print("Vui lòng nhập số trong khoảng 1–9.")
```

```
n = int(input("Nhập số n: "))
if n <= 1:
    print(n, "không phải là số nguyên tố")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print(n, "là số nguyên tố")
    else:
        print(n, "không phải là số nguyên tố")
```
